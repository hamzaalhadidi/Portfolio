#### Remove the Loan_ID column because it is a useless identifier for modeling.

```python
df = df.drop(['Loan_ID'], axis=1)
```

```python
df.isnull().sum()*100/len(df)
```

```python
string_columns = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']

for column in string_columns:
    unique_values = df[column].unique()
    print(f'The Unique value in {column}: {unique_values}')
```

#### Fill in the empty values for the number columns

```python
df["LoanAmount"] = df["LoanAmount"].fillna(df["LoanAmount"].median())
df["Loan_Amount_Term"] = df["Loan_Amount_Term"].fillna(df["Loan_Amount_Term"].median())
df["Credit_History"] = df["Credit_History"].fillna(df["Credit_History"].mode()[0])
```

#### Fill in the empty values ​​in the text columns

```python
categorical_cols = [
    "Gender", "Married", "Dependents",
    "Education", "Self_Employed",
    "Property_Area", "Loan_Status"
]

for col in categorical_cols:
    df[col] = df[col].fillna(df[col].mode()[0])
```

```python
df.isnull().sum()*100/len(df)
```

```python
df["Dependents"] = df["Dependents"].replace("3+", "3")
df["Dependents"] = df["Dependents"].fillna(0)  # تحويل القيم الفارغة إلى صفر
df["Dependents"] = df["Dependents"].astype(int)
```

```python
df["Dependents"].unique()
```

```python
df.head(10)
```

#### Save data after cleaning

```python
df.to_csv("loan_cleaned.csv", index=False, encoding='utf-8')
```

#### Converts categorical columns in a DataFrame into binary (0 and 1) numeric columns for use in machine learning models

```python
df = pd.get_dummies(df, columns=['Gender', 'Married', 'Education', 'Self_Employed', 'Property_Area'], drop_first=True)
```

```python
print(df.columns)
```

#### Convert column 'loan Status' from (Y/N) to (1/0)

```python
df["Loan_Status"] = df["Loan_Status"].map({"Y": 1, "N": 0})
```

```python
df['Loan_Status'] = df['Loan_Status'].astype(int)
```

#### Converts all boolean (True/False) columns in the DataFrame to integers (1/0) for easier use in computational models

```python
bool_cols = df.select_dtypes(include="bool").columns
df[bool_cols] = df[bool_cols].astype(int)
```

```python
ordered_columns = [
    "Gender_Male",
    "Married_Yes",
    "Dependents",
    "Education_Not Graduate",
    "Self_Employed_Yes",
    "ApplicantIncome",
    "CoapplicantIncome",
    "LoanAmount",
    "Loan_Amount_Term",
    "Credit_History",
    "Property_Area_Semiurban",
    "Property_Area_Urban",
    "Loan_Status"
]

df = df[ordered_columns]
```
