#### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use("default")
sns.set_theme(style="whitegrid")
```

#### Load Dataset

```python
df = pd.read_csv("loan_cleaned.csv")
```

#### Loan status distribution heatmap

```python
plt.figure(figsize=(18, 7))
numeric_df = df.select_dtypes(include=["int64", "float64"])
sns.heatmap(numeric_df.corr(), annot=True, cmap="RdYlGn", fmt=".2f")
plt.title("Correlation Heatmap (Numerical Features)", fontsize=16)
plt.show()
```

#### Loan status breakdown

```python
sns.countplot(x="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.xticks([0, 1], ["Loan Approved", "Loan Rejected"])
plt.xlabel("Loan Status", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Loan Status Distribution")
plt.show()
```

#### Gender vs Loan Status

```python
sns.countplot(x="Gender", hue="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.legend(labels=["Approved", "Rejected"])
plt.xlabel("Gender", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Gender vs Loan Status")
plt.show()
```

#### Married vs Loan Status

```python
sns.countplot(x="Married", hue="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.legend(labels=["Approved", "Rejected"])
plt.xlabel("Married", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Marital Status vs Loan Status")
plt.show()
```

#### Education vs Loan Status

```python
sns.countplot(x="Education", hue="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.legend(labels=["Approved", "Rejected"])
plt.xlabel("Education", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Education Level vs Loan Status")
plt.show()
```

#### Self Employed vs Loan Status

```python
sns.countplot(x="Self_Employed", hue="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.legend(labels=["Approved", "Rejected"])
plt.xlabel("Self Employed", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Self Employment vs Loan Status")
plt.show()
```

#### Property Area vs Loan Status

```python
sns.countplot(x="Property_Area", hue="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.legend(labels=["Approved", "Rejected"])
plt.xlabel("Property Area", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.title("Property Area vs Loan Status")
plt.show()
```

#### Distribution of Applicant Income

```python
sns.histplot(df["ApplicantIncome"], bins=40, kde=True)
plt.xlabel("Applicant Income")
plt.ylabel("Frequency")
plt.title("Distribution of Applicant Income")
plt.show()
```

#### Dependents vs Loan Status

```python
sns.countplot(x="Dependents", hue="Loan_Status", data=df, palette=["#1f77b4", "#ff7f0e"])
plt.legend(labels=["Approved", "Rejected"])
plt.xlabel("Number of Dependents")
plt.ylabel("Count")
plt.title("Dependents vs Loan Status")
plt.show()
```

#### Coapplicant Income Distribution

```python
sns.histplot(df["CoapplicantIncome"], bins=40, kde=True)
plt.xlabel("Coapplicant Income")
plt.ylabel("Frequency")
plt.title("Distribution of Coapplicant Income")
plt.show()
```

#### Applicant Income by Gender (Density)

```python
g = sns.FacetGrid(df, hue="Gender", aspect=4)
g.map(sns.kdeplot, "ApplicantIncome", fill=True)
g.add_legend()
plt.xlabel("Applicant Income", fontsize=14)
plt.ylabel("Density", fontsize=14)
plt.title("Applicant Income Distribution by Gender")
plt.show()
```
